```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [12, 8]
```

```python
with open ('../data/map_data.txt') as f:
    map_df = pd.read_table(f, sep='\t', header=None, names=['x_f','y_f','id'], lineterminator='\n')

with open ('../data/gt_data.txt') as f:
    gt_df = pd.read_table(f, sep=' ', header=None, names=['x','y','theta'], lineterminator='\n')

with open ('../data/control_data.txt') as f:
    measurement_df = pd.read_table(f, sep=' ', header=None, names=['velocity','yawrate'], lineterminator='\n')
```

```python
plt.figure()
map_df.plot(kind='scatter', x='x_f', y='y_f')
gt_df.plot(kind='scatter', x='x', y='y')
measurement_df.plot()
```
